```python
import yfinance as yf
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt

# دانلود داده‌های سهام AAPL
ticker = "AAPL"
data = yf.download(ticker, start="2020-01-01", end="2024-09-27")  # تاریخ پایان به روز شده

# انتخاب ستون 'Close' به عنوان متغیر وابسته
data = data[['Close']].dropna()

# ایجاد متغیرهای مستقل (روزهای معاملاتی)
data['Date'] = np.arange(len(data))

# تقسیم داده‌ها به مجموعه‌های آموزش و تست
train_data = data[data.index < '2023-01-01']
test_data = data[data.index >= '2023-01-01']

# ایجاد متغیرهای مستقل و وابسته برای مجموعه آموزش
X_train = train_data[['Date']]
y_train = train_data['Close']

# ایجاد متغیرهای مستقل و وابسته برای مجموعه تست
X_test = test_data[['Date']]
y_test = test_data['Close']

# ایجاد مدل رگرسیون خطی
model = LinearRegression()

# آموزش مدل با استفاده از داده‌ها
model.fit(X_train, y_train)

# پیش‌بینی مقادیر با استفاده از مدل برای مجموعه آموزش
y_train_pred = model.predict(X_train)

# پیش‌بینی مقادیر با استفاده از مدل برای مجموعه تست
y_test_pred = model.predict(X_test)

# محاسبه معیار R² برای مجموعه آموزش
r2_train = r2_score(y_train, y_train_pred)
print(f"Train R²: {r2_train}")

# محاسبه معیار R² برای مجموعه تست
r2_test = r2_score(y_test, y_test_pred)
print(f"Test R²: {r2_test}")

# رسم نمودار داده‌های واقعی و پیش‌بینی شده برای مجموعه آموزش
plt.figure(figsize=(10, 6))
plt.plot(train_data['Date'], y_train, label='Actual Train')
plt.plot(train_data['Date'], y_train_pred, label='Predicted Train', linestyle='--')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('Linear Regression on AAPL Stock Data (Train)')
plt.legend()
plt.savefig('linear_regression_train.png')
plt.show()

# رسم نمودار داده‌های واقعی و پیش‌بینی شده برای مجموعه تست
plt.figure(figsize=(10, 6))
plt.plot(test_data['Date'], y_test, label='Actual Test')
plt.plot(test_data['Date'], y_test_pred, label='Predicted Test', linestyle='--')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('Linear Regression on AAPL Stock Data (Test)')
plt.legend()
plt.savefig('linear_regression_test.png')
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt

# دانلود داده‌های سهام AAPL
ticker = "AAPL"
data = yf.download(ticker, start="2020-01-01", end="2024-09-27")  # تاریخ پایان به روز شده

# انتخاب ستون 'Close' به عنوان متغیر وابسته
data = data[['Close']].dropna()

# ایجاد متغیرهای مستقل (روزهای معاملاتی)
data['Date'] = np.arange(len(data))

# تقسیم داده‌ها به مجموعه‌های آموزش و تست
train_data = data[data.index < '2023-01-01']
test_data = data[data.index >= '2023-01-01']

# ایجاد متغیرهای مستقل و وابسته برای مجموعه آموزش
X_train = train_data[['Date']]
y_train = train_data['Close']

# ایجاد متغیرهای مستقل و وابسته برای مجموعه تست
X_test = test_data[['Date']]
y_test = test_data['Close']

# ایجاد ویژگی‌های پولینومیال برای مجموعه آموزش
poly = PolynomialFeatures(degree=3)  # درجه پولینومیال را می‌توانید تغییر دهید
X_poly_train = poly.fit_transform(X_train)

# ایجاد مدل رگرسیون خطی
model = LinearRegression()

# آموزش مدل با استفاده از داده‌های پولینومیال
model.fit(X_poly_train, y_train)

# پیش‌بینی مقادیر با استفاده از مدل برای مجموعه آموزش
y_train_pred = model.predict(X_poly_train)

# ایجاد ویژگی‌های پولینومیال برای مجموعه تست
X_poly_test = poly.fit_transform(X_test)

# پیش‌بینی مقادیر با استفاده از مدل برای مجموعه تست
y_test_pred = model.predict(X_poly_test)

# محاسبه معیار R² برای مجموعه آموزش
r2_train = r2_score(y_train, y_train_pred)
print(f"Train R²: {r2_train}")

# محاسبه معیار R² برای مجموعه تست
r2_test = r2_score(y_test, y_test_pred)
print(f"Test R²: {r2_test}")

# رسم نمودار داده‌های واقعی و پیش‌بینی شده برای مجموعه آموزش
plt.figure(figsize=(10, 6))
plt.plot(train_data['Date'], y_train, label='Actual Train')
plt.plot(train_data['Date'], y_train_pred, label='Predicted Train', linestyle='--')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('Polynomial Regression on AAPL Stock Data (Train)')
plt.legend()
plt.savefig('polynomial_regression_train.png')
plt.show()

# رسم نمودار داده‌های واقعی و پیش‌بینی شده برای مجموعه تست
plt.figure(figsize=(10, 6))
plt.plot(test_data['Date'], y_test, label='Actual Test')
plt.plot(test_data['Date'], y_test_pred, label='Predicted Test', linestyle='--')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('Polynomial Regression on AAPL Stock Data (Test)')
plt.legend()
plt.savefig('polynomial_regression_test.png')
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt

# دانلود داده‌های سهام AAPL
ticker = "AAPL"
data = yf.download(ticker, start="2020-01-01", end="2024-09-27")

# انتخاب ستون‌های مورد نیاز به عنوان متغیرهای مستقل و وابسته
data = data[['Open', 'High', 'Low', 'Close', 'Volume']].dropna()

# تقسیم داده‌ها به مجموعه‌های آموزش و تست
train_data = data[data.index < '2023-01-01']
test_data = data[data.index >= '2023-01-01']

# ایجاد متغیرهای مستقل و وابسته برای مجموعه آموزش
X_train = train_data[['Open', 'High', 'Low', 'Volume']]
y_train = train_data['Close']

# ایجاد متغیرهای مستقل و وابسته برای مجموعه تست
X_test = test_data[['Open', 'High', 'Low', 'Volume']]
y_test = test_data['Close']

# ایجاد وزن‌ها برای مجموعه آموزش (به عنوان مثال، وزن‌ها می‌توانند به صورت تصادفی یا بر اساس معیار خاصی تعیین شوند)
weights = np.random.rand(len(y_train))

# ایجاد مدل رگرسیون وزن‌دار خطی
model = LinearRegression()

# آموزش مدل با استفاده از داده‌ها و وزن‌ها
model.fit(X_train, y_train, sample_weight=weights)

# پیش‌بینی مقادیر با استفاده از مدل برای مجموعه آموزش
y_train_pred = model.predict(X_train)

# پیش‌بینی مقادیر با استفاده از مدل برای مجموعه تست
y_test_pred = model.predict(X_test)

# محاسبه معیار R² برای مجموعه آموزش
r2_train = r2_score(y_train, y_train_pred)
print(f"Train R²: {r2_train}")

# محاسبه معیار R² برای مجموعه تست
r2_test = r2_score(y_test, y_test_pred)
print(f"Test R²: {r2_test}")

# رسم نمودار داده‌های واقعی و پیش‌بینی شده برای مجموعه آموزش
plt.figure(figsize=(10, 6))
plt.plot(train_data.index, y_train, label='Actual Train')
plt.plot(train_data.index, y_train_pred, label='Predicted Train', linestyle='--')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('Weighted Linear Regression on AAPL Stock Data (Train)')
plt.legend()
plt.savefig('weighted_linear_regression_train.png')
plt.show()

# رسم نمودار داده‌های واقعی و پیش‌بینی شده برای مجموعه تست
plt.figure(figsize=(10, 6))
plt.plot(test_data.index, y_test, label='Actual Test')
plt.plot(test_data.index, y_test_pred, label='Predicted Test', linestyle='--')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('Weighted Linear Regression on AAPL Stock Data (Test)')
plt.legend()
plt.savefig('weighted_linear_regression_test.png')
plt.show()
```
